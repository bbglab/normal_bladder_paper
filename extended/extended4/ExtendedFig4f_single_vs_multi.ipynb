# Load all methods data

```python
gene_order = panel_no_hotspots
```

### Omega

```python
omega_file = f"{deepcsa_run_dir}/omega/output_mle.all_samples.tsv"
omega_data = pd.read_table(omega_file)
omega_data = omega_data[omega_data["impact"].isin(['missense', 'truncating'])]
omega_data = omega_data[omega_data["gene"].isin(gene_order)].reset_index(drop = True)
omega_data.head(10)
```

```python
omega_truncating = omega_data[omega_data["impact"] == "truncating"].reset_index(drop = True)[["gene", "dnds", "pvalue", "lower", "upper"]]
omega_truncating.columns = ["GENE", "omega_trunc", "pvalue", "lower", "upper"]


omega_missense = omega_data[omega_data["impact"] == "missense"].reset_index(drop = True)[["gene", "dnds", "pvalue", "lower", "upper"]]
omega_missense.columns = ["GENE", "omega_mis", "pvalue", "lower", "upper"]
omega_missense
```

### Omega multi

```python
omega_file = f"{deepcsa_run_dir}/omega/output_mle.all_samples.multi.tsv"
omega_data_multi = pd.read_table(omega_file)
omega_data_multi = omega_data_multi[omega_data_multi["impact"].isin(['missense', 'truncating'])]
omega_data_multi = omega_data_multi[omega_data_multi["gene"].isin(gene_order)].reset_index(drop = True)
omega_data_multi.head(10)
```

```python
omega_truncating_multi = omega_data_multi[omega_data_multi["impact"] == "truncating"].reset_index(drop = True)[["gene", "dnds", "pvalue", "lower", "upper"]]
omega_truncating_multi.columns = ["GENE", "omega_trunc", "pvalue", "lower", "upper"]


omega_missense_multi = omega_data_multi[omega_data_multi["impact"] == "missense"].reset_index(drop = True)[["gene", "dnds", "pvalue", "lower", "upper"]]
omega_missense_multi.columns = ["GENE", "omega_mis", "pvalue", "lower", "upper"]
omega_missense_multi
```

```python
omega_missense_multi['multi'] = True
omega_missense['multi'] = False
omega_truncating_multi['multi'] = True
omega_truncating['multi'] = False
```

```python
omega_missense_all = pd.concat((omega_missense, omega_missense_multi))
omega_truncating_all = pd.concat((omega_truncating, omega_truncating_multi))
omega_missense_all
```

```python
gene_order = list(omega_truncating.sort_values("omega_trunc", ascending= False)["GENE"].values)
```

```python
pvalue_thres = 0.05
```

```python
# Create the figure and subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(3, 1.78), gridspec_kw={'height_ratios': [omega_truncating_all['omega_trunc'].max(),
                                                                                    omega_missense_all['omega_mis'].max()
                                                                                   ]})

# Separate data based on significance of p-value
omega_truncating_sig = omega_truncating_all[omega_truncating_all["pvalue"] <= pvalue_thres].reset_index(drop = True)
omega_truncating_notsig = omega_truncating_all[omega_truncating_all["pvalue"] > pvalue_thres].reset_index(drop = True)

# Plot the second bar plot in the middle subplot
sns.barplot(data=omega_truncating_notsig, x='GENE', y='omega_trunc',
            ax=ax1, alpha=1,
            fill = False,
            hue = 'multi',
            hue_order = [False, True],
            # legend = False,
            order = gene_order,
            palette = {False : metrics_colors_dictionary["omega_trunc"],
                       True : metrics_colors_dictionary["omega_trunc"]
                      }
                       )


# Plot the second bar plot in the middle subplot
sns.barplot(data=omega_truncating_sig, x='GENE', y='omega_trunc',
            ax=ax1, alpha=1,
            legend = False,
            hue = 'multi',
            hue_order = [False, True],
            order = gene_order,
            palette = {False : metrics_colors_dictionary["omega_trunc"],
                       True : metrics_colors_dictionary["omega_trunc"]
                      },
            # color = metrics_colors_dictionary["omega_trunc"],
            edgecolor = None
           )

sns.barplot(data=omega_truncating_all[omega_truncating_all['multi'] == True], x='GENE', y='omega_trunc',
            ax=ax1, alpha=1,
            fill = False,
            hue = 'multi',
            hue_order = [False, True],
            order = gene_order,
            palette = {False : metrics_colors_dictionary["omega_trunc"],
                       True : 'black'
                      },
            dodge = True,
            edgecolor = 'black',
            hatch = "//////",
            linewidth = 0.5
           )


ax1.set_xlabel('')
ax1.set_ylabel('dN/dS\ntruncating',
                  #fontsize = 12,
                  rotation = 0, verticalalignment = 'center',
               horizontalalignment = 'right')
# ax1.set_yticks([0,10,20,30,40,50])
# ax1.set_yticklabels([0,10,20,30,40,50])
ax1.set_xticklabels([])  # Hide x-axis labels on the middle plot
ax1.axhline(1, color='black', linestyle='--', linewidth = 0.5)

# Remove top and right spines
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.legend(frameon = False)



# Separate data based on significance of p-value
omega_missense_sig = omega_missense_all[omega_missense_all["pvalue"] <= pvalue_thres].reset_index(drop = True)
omega_missense_notsig = omega_missense_all[omega_missense_all["pvalue"] > pvalue_thres].reset_index(drop = True)

# Plot the second bar plot in the middle subplot
sns.barplot(data=omega_missense_notsig, x='GENE', y='omega_mis',
            ax=ax2, alpha=1,
            fill = False,
            hue = 'multi',
            hue_order = [False, True],
            legend = False,
            order = gene_order,
            palette = {False : metrics_colors_dictionary["omega_miss"],
                       True : metrics_colors_dictionary["omega_miss"]
                      }
           )


# Plot the second bar plot in the middle subplot
sns.barplot(data=omega_missense_sig, x='GENE', y='omega_mis',
            ax=ax2, alpha=1,
            hue = 'multi',
            hue_order = [False, True],
            legend = False,
            order = gene_order,
            palette = {False : metrics_colors_dictionary["omega_miss"],
                       True : metrics_colors_dictionary["omega_miss"]
                      },
            edgecolor = None
           )



sns.barplot(data=omega_missense_all[omega_missense_all['multi'] == True], x='GENE', y='omega_mis',
            ax=ax2, alpha=1,
            fill = False,
            hue = 'multi',
            hue_order = [False, True],
            order = gene_order,
            palette = {False : metrics_colors_dictionary["omega_miss"],
                       True : 'black'
                      },
            legend = False,
            dodge = True,
            edgecolor = 'black',
            hatch = "//////",
            linewidth = 0.5
           )

ax2.set_xlabel('')
ax2.set_ylabel('dN/dS\nmissense',
                  #fontsize = 12,
                  rotation = 0, verticalalignment = 'center',
               horizontalalignment = 'right')
# ax2.set_yticks([0,3,6])
# ax2.set_yticklabels([0,3,6])

ax2.set_xticklabels([])  # Hide x-axis labels on the upper plot
ax2.axhline(1, color='black', linestyle='--', linewidth = 0.5)
# Remove top and right spines
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)

ax2.set_xticks(range(len(gene_order)))
ax2.set_xticklabels(gene_order, rotation=90)



# Custom legend
legend_elements = [
    Patch(facecolor=metrics_colors_dictionary["omega_trunc"], edgecolor=None,
          # label="Expanded clones"
          label="Multiple mutations in the same\nsample accounted only once"
         ),
    Patch(facecolor=metrics_colors_dictionary["omega_trunc"], edgecolor="black", hatch="//////",
          # label="Independent mutations"
          label="Multiple mutations in the same\nsample accounted as many\ntimes as they are observed",
          linewidth = 0.5
         )
]
ax1.legend(handles=legend_elements, frameon=False, loc="upper right")


plt.tight_layout()

plt.savefig(f"plots/ExtendedFig4f.pdf", bbox_inches='tight', dpi = 300)

plt.show()
```
